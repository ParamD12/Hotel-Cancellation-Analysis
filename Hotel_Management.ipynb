```python
# The 'reservation_status_date' column is of object datatype, 
# so lets convert it to datetime

df['reservation_status_date'] = pd.to_datetime(df['reservation_status_date'], dayfirst=True)
```

```python
# The Columns where object data type is present are 
# the ones to hold categorical values, 
# and we can find the distinct number of categories in these columns
# To get the summary statistics of columns with object datatype

df.describe(include = 'object') 
```

```python
# To find the categories in this column

for col in df.describe(include = 'object').columns:
    print(col)
    print(df[col].unique())
    print('-'*50)
```

```python
# Check Missing Values

df.isnull().sum()
```

```python
# The number of null values for agent and company column is large
# We will drop these rows
# We don't need agent, and companies either

df.drop(['company', 'agent'], axis = 1, inplace = True)
```

```python
# For other missing values, like for the country column
# We can drop the rows with missing values

df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
# To see summary statistics of numerical columns

df.describe()
```

```python
# Now certainly there are some outliers, but either they don't affect our analysis, or safe-side we can remove those as well
# We earlier assumed that there are no outliers in the dataset

# But there is this one significant outlier in adr column - which shows average daily rate of hotel room

df['adr'].plot(kind='box')
```

```python
# To remove this outlier 

df = df[df['adr']<5000]
```

```python
df['adr'].plot(kind='box')
```

# Data Analysis and Visualisation

```python
# To find the times and percentage of cancellations being made

cancelled_percentage = df['is_canceled'].value_counts(normalize=True)

cancelled_percentage
```

```python
# Visualising the cancellaion

print(cancelled_percentage*100)
plt.figure(figsize=(7,6))
plt.title('Reservation Status Count')
plt.bar(['Not Cancelled', 'Cancelled'], df['is_canceled'].value_counts(), edgecolor = 'k', width = 0.7)
plt.show

print('-'*60)
print(df['is_canceled'].value_counts())
```

```python
# Now we'll check whose cancellations are more - Resort Hotels or City Hotels

plt.figure(figsize=(8,4))

ax1 = sns.countplot(x = 'hotel', hue = 'is_canceled', data = df, palette = 'Blues')
legend_labels,_ = ax1.get_legend_handles_labels()
ax1.legend(bbox_to_anchor = (1,1))

plt.title('Reservation Status in Different Hotels', size = 20)
plt.xlabel('Hotel')
plt.ylabel('Number of Reseravtions')

plt.show()
print('0 - > Not Cancelled')
print('1 - > Cancelled')
```

```python
# To find percentage of cancellations in Resort Hotel and City Hotel

# In Resort Hotels ADR can be the main factor
# While in City Hotels - it can be maintanence or other reasons

resort_hotel = df[df['hotel']  == 'Resort Hotel']

resort_hotel['is_canceled'].value_counts(normalize=True)*100
```

```python
city_hotel = df[df['hotel']  == 'City Hotel']

city_hotel['is_canceled'].value_counts(normalize=True)*100
```

```python
# Now we'll check the effect of price of Hotel Room or ADR on cancellation

# We'll find average reservations per date

resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
# Creating a Visualisation for Average Daily Rate 

plt.figure(figsize=(20,8))
plt.title('Average Daily Rate in City and Resort Hotel', size = 25)

plt.plot(resort_hotel.index, resort_hotel['adr'], label = 'Resort Hotel')
plt.plot(city_hotel.index, city_hotel['adr'], label = 'City Hotel')

plt.legend(fontsize = 15)
plt.show()
```

```python
# From here we verify our hypothesis that ADR is one of the reasons in more frequent cancellations in Resort Hotels
# As we can clearly see that ADR of Resort Hotels is significantly more than City Hotels
```

```python
# To find the Months where most reservations and cancellations are made

df['month'] = df['reservation_status_date'].dt.month

plt.figure(figsize=(16,8))
ax1 = sns.countplot(x='month', data=df, hue='is_canceled', palette='bright')

legend_labels, _ = ax1.get_legend_handles_labels()
ax1.legend(bbox_to_anchor = (1,1))

plt.title('Reservation Status Per Month', size = 20)
plt.xlabel('Month')
plt.ylabel('Number of Reseravtions')
plt.legend(['not_cancelled', 'cancelled'])

plt.show()
```

```python
# By this we can infer that the most cancellations are made in December-January
# and the least cancellations are made in July-September.

# Now lets check price for each month
```

```python
plt.figure(figsize=(15,8))

plt.title('ADR Per Month', fontsize = 20)

sns.barplot(x = 'month', y = 'adr', data = df[df['is_canceled'] == 1].groupby('month')[['adr']].sum().reset_index())

plt.show()
```

```python
# This as well proves our Hypothesis that when ADR is high, Cancellation rates are higher
```

```python
# Now we'll check the cancellation rates per each country for Top 10 Countries

cancelled_data = df[df['is_canceled'] == 1]
top_10_country = cancelled_data['country'].value_counts()[:10]

plt.figure(figsize=(10,10))
plt.title('Top 10 Countries with Cancelled Reservation', fontsize = 20)
plt.pie(top_10_country, autopct = '%.2f', labels = top_10_country.index)

plt.show()
```

```python
# Now by the Chart above we need to give suitable suggestions to Hotels in Portugal 
# to give more promotional discounts, provide better services, do better marketing etc.
```

```python
# Now we'll find about the source where clients are coming from - Online Travel Agents or Offline

df['market_segment'].value_counts()
```

```python
# To see percentage count of this 

df['market_segment'].value_counts(normalize=True)*100
```

```python
# This shows us that more clients are acqired from Online Travel Agencies
```

```python
# Now we'll check for the same dynamics but for the cancelled reservations

cancelled_data['market_segment'].value_counts(normalize=True)*100
```

```python
# This also shows that Online Travel Agents have more cancellation frequency 
```

```python
# Now we'll compare the ADR of Cancelled Reservations and Non-Cancelled Reservations

cancelled_df_adr = cancelled_data.groupby('reservation_status_date')[['adr']].mean()
cancelled_df_adr.reset_index(inplace = True)
cancelled_df_adr.sort_values('reservation_status_date', inplace = True)

not_cancelled_data = df[df['is_canceled'] == 0]
not_cancelled_df_adr = not_cancelled_data.groupby('reservation_status_date')[['adr']].mean()
not_cancelled_df_adr.reset_index(inplace = True)
not_cancelled_df_adr.sort_values('reservation_status_date', inplace = True)


plt.figure(figsize = (20,8))
plt.title('Average Daily Rate', fontsize = 20)
plt.plot(not_cancelled_df_adr['reservation_status_date'], not_cancelled_df_adr['adr'], label = 'not cancelled')
plt.plot(cancelled_df_adr['reservation_status_date'], cancelled_df_adr['adr'], label = 'cancelled')
plt.legend(fontsize = 20)
plt.show
```

```python
# Since there is inconsistency in dataset we'll filter and plot data from 2016 - Sept, 2017

cancelled_df_adr = cancelled_df_adr[(cancelled_df_adr['reservation_status_date']>'2016') & (cancelled_df_adr['reservation_status_date']<'2017-09')]
not_cancelled_df_adr = not_cancelled_df_adr[(not_cancelled_df_adr['reservation_status_date']>'2016') & (not_cancelled_df_adr['reservation_status_date']<'2017-09')]
```

```python
# We'll plot data again

plt.figure(figsize = (20,8))
plt.title('Average Daily Rate', fontsize = 20)
plt.plot(not_cancelled_df_adr['reservation_status_date'], not_cancelled_df_adr['adr'], label = 'not cancelled')
plt.plot(cancelled_df_adr['reservation_status_date'], cancelled_df_adr['adr'], label = 'cancelled')
plt.legend(fontsize = 20)
plt.show
```

```python
# By this we can conclude that ADR is the main varibale affecting cancellation rates
```
